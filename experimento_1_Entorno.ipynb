```python
import subprocess
import sys
```

```python
# 1. Desinstalar versiones actuales
subprocess.run([sys.executable, '-m', 'pip', 'uninstall', 'torch', 'torchvision', 'torchaudio', '-y'],
               capture_output=True)
```

```python
# Instalar PyTorch con GPU
result = subprocess.run([sys.executable, '-m', 'pip', 'install', 'torch', 'torchvision'],
                       capture_output=True, text=True)
```

```python
print("Resultado instalación:")
print(f"Return code: {result.returncode}")
if result.returncode == 0:
    print("✓ Instalación exitosa")
else:
    print("✗ Error en instalación")
    print(result.stderr)

print("\nREINICIA EL RUNTIME y luego ejecuta:")
print("""
import torch
print(f"PyTorch: {torch.__version__}")
print(f"GPU: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print("✓ Listo para MaskDINO")
""")
```
